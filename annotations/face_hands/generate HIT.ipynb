```python
import os
import pandas as pd
```

```python
def print_trial(trial_num):
    print('''<crowd-image-classifier  
        src="${img_src'''+ trial_num + '''}"
        categories="['Face and Hand', 'Face only', 'Hand only','Neither']"
        header="Does the image contain a face or a hand?"
        name="trial-'''+ trial_num + '''">
        <short-instructions>
        <full-instructions header="Does this image contain a face or a hand?">
            <h2>Detailed instructions</h2>
            <li>Begin by carefully examining the entire image</li>
             <li>Then determine if a human face and/or a hand occurs in the image, and select the correct option. </li>
            </ol>
        </full-instructions>
    </crowd-image-classifier>''')
```

```python
def make_trials():
    for trial_num in range(1,101):
        print_trial(str(trial_num))
```

```python
make_trials()
```
