```python
from detector import FaceDetector
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import os
```

```python
det = FaceDetector('mtcnn') 
#constructor auto-loads the appropriate detector
```

```python
#choose an "easy" image that all detectors should get for sanity check
img_path = '/scratch/users/agrawalk/toy/family.jpg'
img = cv2.imread(img_path)
```

```python
faces = det.detect_faces(img)
if len(faces) > 0:
    print(faces)
    print(type(faces[0]))
#detect_faces takes in an integer, and 
```

```python
#Ensuring the bounding boxes make sense

fig,ax = plt.subplots(1)

# Display the image
ax.imshow(img)
for bb in faces:
    x, y, w, h = bb
    rect = patches.Rectangle((x, y), w, h, linewidth=1, edgecolor='r', facecolor='none')
    # Add the patch to the Axes
    ax.add_patch(rect)
plt.show()
```
